```python
import pandas as pd
import numpy as np
from pylab import rcParams
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Seasons_Stats.csv')
# data.info()
# # data.isin('Player' == 'Curly Armstrong')
player_data = data[['Year','Player','Tm','Pos','Age','G','3PA','3P%','2PA','2P%','PF','PTS']]
p_jk = player_data.loc[player_data['Player'].isin(['Michael Jordan*','Kobe Bryant'])]
p_jk = p_jk.loc[ (p_jk['Age'] >= 21) & (p_jk['Age'] <= 29) ]


```

```python
p_jk_3 = p_jk.pivot_table(
                        index='Age', 
                        columns='Player', 
                        values='3P%').fillna(0).applymap(float)
p_jk_2 = p_jk.pivot_table(
                        index='Age', 
                        columns='Player', 
                        values='2P%').fillna(0).applymap(float)
p_jk_all = p_jk.pivot_table(
                        index='Age', 
                        columns='Player', 
                        values='PTS').fillna(0).applymap(float)

p_jk_pf = p_jk.pivot_table(
                        index='Age', 
                        columns='Player', 
                        values='PF').fillna(0).applymap(float)
```

```python
p_jk_3.plot()
plt.title('3-Point Field Goals Percentage')


# При стравнении статистики игр Джордона и Браента, видно что процент реализации трех очковых бросков у Браента выше,
# чем у Джордона в аналогичном возрасте. В целом у Браента статиска 3х очковых бросков более стабильна, чем у Джордана.
```

```python
p_jk_2.plot()
plt.title('2-Point Field Goals Percentage')

# НА данном графике мы видим, что реализации 2х очковы бросков у Джордана выше, имеется сильная просадка по точности в период
# с 21 по 22 года. На это стоило бы обратить внимание, если не знать что в данный период Джордан уходил ищ НБА в связи со смертью отца.
```

```python
p_jk_all.plot()
plt.title('Point Per Games')

# На данном графике видим, что пик по общему кол-ву набранных очков за весь период у Джордана пришелся на 23 года и постепенно
# с годами снижался. Браент наооборот с годами улучшал статистику и нарашивал кол-во бросков.
```

```python
p_jk_pf.plot.bar()
plt.title('Personal Fouls')

# Джордан показывает хорошую динамику на снижение по кол-во фолов получаемых в сезоне.
# Браент в свою очередь получает примерно одинаковое кол-во фолов в каждом годе. 
```

```python
p_jk.groupby('Player').mean().plot.pie(y='PTS')

# По общему кол-ву бросков Браент незначительно уступает Джордану.
# В целом рассмотрев статистику данных игроков в период их жизни с 21 по 29 лет включительно, можно сделать ряд выводов:
# В период с 26 по 27 лет , примерно в 26.5 эти два игрока играли примерно одинаково хорошо. У них юыло примерно равный процент
# попаданий 3х очковых бросков, примерно одинакове кол-во полученных фолов, и общее кол-во забитых очков было в районе 2600 очков.
# Видно что в своие 27 лет Джордан почти на 10% лучше реализаовывал 2х очковые броски, чем Браент. Но при одинаковом  общем кол-ве
# очков и остальной статиске, можно предположить что, Браент набрал тоже самое кол-во за счет большего кол-ва реализованных 3х очковых
```
