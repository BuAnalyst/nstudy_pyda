```python
import pandas as pd
```

```python
data = pd.read_csv('ml-latest-small/ratings.csv')

data_max = data[['userId','timestamp']]
data_min = data[['userId','timestamp']]

data_max = data_max.groupby('userId').max()
data_min = data_min.groupby('userId').min()

data = data.groupby('userId').count()
data = data.loc[data['rating'] >= 100].reset_index()
data = data[['userId','rating']]

data = (data
        .merge(data_max, how='left', on ='userId') 
        .merge(data_min, how='left', on ='userId')
        [['userId','timestamp_x','timestamp_y']]
       )

data['live_time_in_hours'] = round((data.timestamp_x - data.timestamp_y) / 3600,0)
data[['userId','live_time_in_hours']].sort_values('live_time_in_hours', ascending=False).head()
```
